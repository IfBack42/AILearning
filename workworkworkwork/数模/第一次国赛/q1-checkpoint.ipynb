```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
NIPT数据分析 - 问题1：Y染色体浓度与孕周、BMI等指标的关系模型

这个脚本实现了完整的Y染色体浓度关系模型分析，包括：
1. 数据读取与预处理
2. 探索性数据分析(EDA)
3. 相关性分析
4. 多元回归模型建立
5. 模型验证与显著性检验
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

class NIPTAnalyzer:
    def __init__(self):
        self.data = None
        self.processed_data = None
        self.model_results = {}
        
    def load_data(self, file_path):
        """读取Excel数据文件"""
        print("=== 步骤1：数据读取 ===")
        try:
            self.data = pd.read_excel('./data/附件.xlsx')
            print(f"成功读取数据，共{len(self.data)}条记录")
            print(f"数据列名：{list(self.data.columns[:10])}...")
            return True
        except Exception as e:
            print(f"数据读取失败：{e}")
            return False
    
    def preprocess_data(self):
        """数据预处理"""
        print("\n=== 步骤2：数据预处理 ===")
        
        # 创建工作副本
        df = self.data.copy()
        
        # 1. 解析孕周格式
        def parse_gestational_week(week_str):
            """将'11w+6'格式转换为数值"""
            if pd.isna(week_str):
                return None
            try:
                if 'w' in str(week_str):
                    parts = str(week_str).replace('w', '').split('+')
                    weeks = float(parts[0])
                    days = float(parts[1]) if len(parts) > 1 else 0
                    return weeks + days / 7
                else:
                    return float(week_str)
            except:
                return None
        
        # 应用孕周转换
        df['孕周数值'] = df['检测孕周'].apply(parse_gestational_week)
        
        # 2. 确保关键数值列为数值类型
        numeric_columns = ['年龄', '身高', '体重', '孕妇BMI', 'Y染色体浓度']
        for col in numeric_columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        
        # 3. 过滤有效数据
        # 只保留Y染色体浓度、孕周、BMI都不为空且Y浓度大于0的记录
        valid_mask = (
            df['Y染色体浓度'].notna() & 
            (df['Y染色体浓度'] > 0) &
            df['孕周数值'].notna() & 
            df['孕妇BMI'].notna() &
            df['年龄'].notna()
        )
        
        df_clean = df[valid_mask].copy()
        
        # 4. 异常值处理（使用四分位数方法）
        def remove_outliers_iqr(data, column, factor=1.5):
            """使用IQR方法移除异常值"""
            Q1 = data[column].quantile(0.25)
            Q3 = data[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - factor * IQR
            upper_bound = Q3 + factor * IQR
            return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
        
        # 对关键变量移除异常值
        original_len = len(df_clean)
        for col in ['Y染色体浓度', '孕妇BMI', '孕周数值']:
            df_clean = remove_outliers_iqr(df_clean, col)
        
        print(f"原始数据：{len(self.data)}条")
        print(f"有效数据：{len(df[valid_mask])}条")
        print(f"移除异常值后：{len(df_clean)}条")
        print(f"数据利用率：{len(df_clean)/len(self.data)*100:.1f}%")
        
        # 5. 创建衍生变量
        df_clean['BMI分类'] = pd.cut(df_clean['孕妇BMI'], 
                                 bins=[0, 18.5, 25, 30, 35, float('inf')],
                                 labels=['偏瘦', '正常', '超重', '肥胖', '重度肥胖'])
        
        df_clean['Y浓度达标'] = (df_clean['Y染色体浓度'] >= 0.04).astype(int)
        
        # 6. 处理重复测量（同一孕妇多次检测）
        print(f"不同孕妇数量：{df_clean['孕妇代码'].nunique()}人")
        print(f"多次检测的孕妇：{df_clean.groupby('孕妇代码').size().gt(1).sum()}人")
        
        self.processed_data = df_clean
        print("数据预处理完成！\n")
        
        return df_clean
    
    def exploratory_analysis(self):
        """探索性数据分析"""
        print("=== 步骤3：探索性数据分析 ===")
        
        df = self.processed_data
        
        # 描述性统计
        key_vars = ['Y染色体浓度', '孕周数值', '孕妇BMI', '年龄']
        desc_stats = df[key_vars].describe()
        print("关键变量描述性统计：")
        print(desc_stats.round(4))
        
        # Y染色体浓度达标情况
        threshold_stats = df['Y浓度达标'].value_counts()
        reach_rate = threshold_stats[1] / len(df) * 100 if 1 in threshold_stats else 0
        print(f"\nY染色体浓度达标情况：")
        print(f"达标(≥4%)：{threshold_stats.get(1, 0)}例 ({reach_rate:.1f}%)")
        print(f"未达标(<4%)：{threshold_stats.get(0, 0)}例 ({100-reach_rate:.1f}%)")
        
        # 可视化分析
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('NIPT数据探索性分析', fontsize=16, fontweight='bold')
        
        # 1. Y染色体浓度分布
        axes[0,0].hist(df['Y染色体浓度'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
        axes[0,0].axvline(x=0.04, color='red', linestyle='--', linewidth=2, label='4%阈值')
        axes[0,0].set_title('Y染色体浓度分布')
        axes[0,0].set_xlabel('Y染色体浓度')
        axes[0,0].set_ylabel('频数')
        axes[0,0].legend()
        
        # 2. 孕周分布
        axes[0,1].hist(df['孕周数值'], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
        axes[0,1].set_title('孕周分布')
        axes[0,1].set_xlabel('孕周')
        axes[0,1].set_ylabel('频数')
        
        # 3. BMI分布
        axes[0,2].hist(df['孕妇BMI'], bins=30, alpha=0.7, color='orange', edgecolor='black')
        axes[0,2].set_title('BMI分布')
        axes[0,2].set_xlabel('BMI')
        axes[0,2].set_ylabel('频数')
        
        # 4. Y浓度vs孕周散点图
        scatter = axes[1,0].scatter(df['孕周数值'], df['Y染色体浓度'], 
                                  c=df['孕妇BMI'], cmap='viridis', alpha=0.6, s=20)
        axes[1,0].set_title('Y染色体浓度 vs 孕周')
        axes[1,0].set_xlabel('孕周')
        axes[1,0].set_ylabel('Y染色体浓度')
        plt.colorbar(scatter, ax=axes[1,0], label='BMI')
        
        # 5. Y浓度vs BMI散点图
        axes[1,1].scatter(df['孕妇BMI'], df['Y染色体浓度'], alpha=0.6, s=20, color='red')
        axes[1,1].set_title('Y染色体浓度 vs BMI')
        axes[1,1].set_xlabel('BMI')
        axes[1,1].set_ylabel('Y染色体浓度')
        
        # 6. BMI分类箱线图
        df.boxplot(column='Y染色体浓度', by='BMI分类', ax=axes[1,2])
        axes[1,2].set_title('不同BMI分类的Y染色体浓度分布')
        axes[1,2].set_xlabel('BMI分类')
        axes[1,2].set_ylabel('Y染色体浓度')
        
        plt.tight_layout()
        plt.show()
        
        return desc_stats
    
    def correlation_analysis(self):
        """相关性分析"""
        print("=== 步骤4：相关性分析 ===")
        
        df = self.processed_data
        
        # 选择数值变量进行相关性分析
        numeric_vars = ['Y染色体浓度', '孕周数值', '孕妇BMI', '年龄', '身高', '体重']
        corr_data = df[numeric_vars].corr()
        
        print("Pearson相关系数矩阵：")
        print(corr_data.round(4))
        
        # 计算显著性
        def calculate_correlation_significance(data, alpha=0.05):
            """计算相关系数的显著性"""
            corr_results = {}
            for i, var1 in enumerate(numeric_vars):
                for j, var2 in enumerate(numeric_vars):
                    if i < j:  # 避免重复计算
                        r, p = stats.pearsonr(data[var1].dropna(), data[var2].dropna())
                        corr_results[f"{var1} vs {var2}"] = {
                            'correlation': r,
                            'p_value': p,
                            'significant': p < alpha
                        }
            return corr_results
        
        corr_significance = calculate_correlation_significance(df)
        
        print("\n重要相关性分析结果：")
        for pair, result in corr_significance.items():
            if 'Y染色体浓度' in pair:
                print(f"{pair}: r={result['correlation']:.4f}, "
                     f"p={result['p_value']:.4f}, "
                     f"{'显著' if result['significant'] else '不显著'}")
        
        # 绘制相关性热图
        plt.figure(figsize=(12, 10))
        mask = np.triu(np.ones_like(corr_data, dtype=bool))
        sns.heatmap(corr_data, mask=mask, annot=True, cmap='coolwarm', center=0,
                    square=True, fmt='.3f', cbar_kws={"shrink": .8})
        plt.title('变量相关性热图')
        plt.tight_layout()
        plt.show()
        
        return corr_data, corr_significance
    
    def build_regression_models(self):
        """建立回归模型"""
        print("=== 步骤5：回归模型建立 ===")
        
        df = self.processed_data
        
        # 准备数据
        X_columns = ['孕周数值', '孕妇BMI', '年龄']
        X = df[X_columns]
        y = df['Y染色体浓度']
        
        # 模型1：基础线性回归
        print("模型1：多元线性回归")
        model1 = LinearRegression()
        model1.fit(X, y)
        
        # 预测
        y_pred1 = model1.predict(X)
        
        # 计算模型指标
        r2_1 = r2_score(y, y_pred1)
        mse_1 = mean_squared_error(y, y_pred1)
        
        print(f"线性回归模型结果：")
        print(f"截距: {model1.intercept_:.6f}")
        for i, coef in enumerate(model1.coef_):
            print(f"{X_columns[i]}系数: {coef:.6f}")
        print(f"R²: {r2_1:.4f}")
        print(f"MSE: {mse_1:.6f}")
        
        # 模型方程
        equation = f"Y染色体浓度 = {model1.intercept_:.6f}"
        for i, coef in enumerate(model1.coef_):
            equation += f" + ({coef:.6f}) × {X_columns[i]}"
        print(f"\n回归方程：\n{equation}")
        
        # 模型2：多项式回归（考虑非线性关系）
        print("\n模型2：多项式回归（2次）")
        poly_features = PolynomialFeatures(degree=2, include_bias=False)
        X_poly = poly_features.fit_transform(X)
        
        model2 = LinearRegression()
        model2.fit(X_poly, y)
        y_pred2 = model2.predict(X_poly)
        
        r2_2 = r2_score(y, y_pred2)
        mse_2 = mean_squared_error(y, y_pred2)
        
        print(f"多项式回归模型结果：")
        print(f"R²: {r2_2:.4f}")
        print(f"MSE: {mse_2:.6f}")
        
        # 模型3：考虑交互作用
        print("\n模型3：包含交互项的回归")
        X_interaction = df[['孕周数值', '孕妇BMI', '年龄']].copy()
        X_interaction['孕周×BMI'] = X_interaction['孕周数值'] * X_interaction['孕妇BMI']
        X_interaction['孕周²'] = X_interaction['孕周数值'] ** 2
        
        model3 = LinearRegression()
        model3.fit(X_interaction, y)
        y_pred3 = model3.predict(X_interaction)
        
        r2_3 = r2_score(y, y_pred3)
        mse_3 = mean_squared_error(y, y_pred3)
        
        print(f"交互作用模型结果：")
        print(f"截距: {model3.intercept_:.6f}")
        for i, coef in enumerate(model3.coef_):
            print(f"{X_interaction.columns[i]}系数: {coef:.6f}")
        print(f"R²: {r2_3:.4f}")
        print(f"MSE: {mse_3:.6f}")
        
        # 保存模型结果
        self.model_results = {
            'linear': {'model': model1, 'r2': r2_1, 'mse': mse_1, 'y_pred': y_pred1},
            'polynomial': {'model': model2, 'r2': r2_2, 'mse': mse_2, 'y_pred': y_pred2},
            'interaction': {'model': model3, 'r2': r2_3, 'mse': mse_3, 'y_pred': y_pred3, 'X': X_interaction}
        }
        
        return self.model_results
    
    def model_validation(self):
        """模型验证"""
        print("=== 步骤6：模型验证 ===")
        
        df = self.processed_data
        y_actual = df['Y染色体浓度']
        
        # 1. 统计检验
        print("1. 模型显著性检验")
        
        # F检验 - 线性回归模型
        X = df[['孕周数值', '孕妇BMI', '年龄']]
        n = len(y_actual)
        k = X.shape[1]  # 自变量个数
        
        model = self.model_results['linear']['model']
        y_pred = self.model_results['linear']['y_pred']
        
        # 计算F统计量
        SSR = np.sum((y_pred - y_actual.mean())**2)  # 回归平方和
        SSE = np.sum((y_actual - y_pred)**2)        # 残差平方和
        MSR = SSR / k
        MSE = SSE / (n - k - 1)
        F_stat = MSR / MSE
        
        # F分布临界值（α=0.05）
        F_critical = stats.f.ppf(0.95, k, n-k-1)
        
        print(f"F统计量: {F_stat:.4f}")
        print(f"F临界值: {F_critical:.4f}")
        print(f"模型整体显著性: {'显著' if F_stat > F_critical else '不显著'}")
        
        # 2. 系数显著性检验（t检验）
        print("\n2. 系数显著性检验")
        
        # 计算标准误
        X_with_intercept = np.column_stack([np.ones(n), X])
        XTX_inv = np.linalg.inv(X_with_intercept.T @ X_with_intercept)
        std_errors = np.sqrt(np.diag(XTX_inv) * MSE)
        
        coefficients = np.array([model.intercept_] + list(model.coef_))
        t_stats = coefficients / std_errors
        t_critical = stats.t.ppf(0.975, n-k-1)  # 双尾检验，α=0.05
        
        var_names = ['截距'] + ['孕周数值', '孕妇BMI', '年龄']
        for i, (name, coef, se, t_stat) in enumerate(zip(var_names, coefficients, std_errors, t_stats)):
            p_value = 2 * (1 - stats.t.cdf(abs(t_stat), n-k-1))
            print(f"{name}: 系数={coef:.6f}, 标准误={se:.6f}, t值={t_stat:.4f}, "
                 f"p值={p_value:.4f}, {'显著' if abs(t_stat) > t_critical else '不显著'}")
        
        # 3. 残差分析
        print("\n3. 残差分析")
        residuals = y_actual - y_pred
        
        # Durbin-Watson检验（检验残差独立性）
        dw_stat = np.sum(np.diff(residuals)**2) / np.sum(residuals**2)
        print(f"Durbin-Watson统计量: {dw_stat:.4f}")
        print(f"残差独立性: {'良好' if 1.5 <= dw_stat <= 2.5 else '存在问题'}")
        
        # 正态性检验（Shapiro-Wilk检验）
        if len(residuals) <= 5000:  # Shapiro-Wilk检验对样本量有限制
            sw_stat, sw_p = stats.shapiro(residuals)
            print(f"Shapiro-Wilk检验: W={sw_stat:.4f}, p={sw_p:.4f}")
            print(f"残差正态性: {'满足' if sw_p > 0.05 else '不满足'}")
        
        # 绘制验证图
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('模型验证分析', fontsize=16, fontweight='bold')
        
        # 1. 实际vs预测散点图
        axes[0,0].scatter(y_actual, y_pred, alpha=0.6, s=20)
        axes[0,0].plot([y_actual.min(), y_actual.max()], [y_actual.min(), y_actual.max()], 
                      'r--', lw=2, label='完美预测线')
        axes[0,0].set_xlabel('实际Y染色体浓度')
        axes[0,0].set_ylabel('预测Y染色体浓度')
        axes[0,0].set_title(f'预测vs实际 (R²={self.model_results["linear"]["r2"]:.4f})')
        axes[0,0].legend()
        
        # 2. 残差图
        axes[0,1].scatter(y_pred, residuals, alpha=0.6, s=20)
        axes[0,1].axhline(y=0, color='red', linestyle='--')
        axes[0,1].set_xlabel('预测值')
        axes[0,1].set_ylabel('残差')
        axes[0,1].set_title('残差图')
        
        # 3. 残差Q-Q图（正态性检验）
        stats.probplot(residuals, dist="norm", plot=axes[0,2])
        axes[0,2].set_title('残差Q-Q图（正态性检验）')
        
        # 4. 残差直方图
        axes[1,0].hist(residuals, bins=30, alpha=0.7, edgecolor='black')
        axes[1,0].set_xlabel('残差')
        axes[1,0].set_ylabel('频数')
        axes[1,0].set_title('残差分布直方图')
        
        # 5. 模型比较
        models = ['linear', 'polynomial', 'interaction']
        r2_scores = [self.model_results[model]['r2'] for model in models]
        axes[1,1].bar(models, r2_scores, color=['blue', 'green', 'orange'])
        axes[1,1].set_ylabel('R² 值')
        axes[1,1].set_title('不同模型R²比较')
        axes[1,1].set_ylim(0, max(r2_scores) * 1.1)
        
        # 6. 特征重要性（基于系数绝对值）
        feature_importance = np.abs(model.coef_)
        feature_names = ['孕周', 'BMI', '年龄']
        axes[1,2].barh(feature_names, feature_importance)
        axes[1,2].set_xlabel('系数绝对值')
        axes[1,2].set_title('特征重要性')
        
        plt.tight_layout()
        plt.show()
        
        return {
            'f_statistic': F_stat,
            'f_critical': F_critical,
            'coefficients': coefficients,
            'std_errors': std_errors,
            't_statistics': t_stats,
            'dw_statistic': dw_stat,
            'residuals': residuals
        }
    
# 主执行函数
def main():
    """主函数"""
    # 创建分析器实例
    analyzer = NIPTAnalyzer()
    
    # 数据文件路径（请根据实际情况修改）
    file_path = "男胎检测.xlsx"
    
    # 执行完整分析流程
    if analyzer.load_data(file_path):
        analyzer.preprocess_data()
        analyzer.exploratory_analysis()
        analyzer.correlation_analysis()
        analyzer.build_regression_models()
        analyzer.model_validation()
        analyzer.generate_report()
        
        print("\n分析完成！模型结果已保存。")
        return analyzer
    else:
        print("分析失败，请检查数据文件路径！")
        return None

# 运行示例
if __name__ == "__main__":
    # 运行分析
    result = main()
    
```
