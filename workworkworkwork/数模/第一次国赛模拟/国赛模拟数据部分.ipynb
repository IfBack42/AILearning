```python
import pandas as pd
import numpy as np

# 读取数据
data = pd.read_csv('data_4.csv')

# 计算相邻数据的差值，用于识别趋势变化点
data['diff'] = data['主路4的车流量 (Traffic flow on the Main road 4)'].diff().abs()

# 设置一个阈值来识别显著变化点（这里使用标准差作为阈值）
threshold = data['diff'].std() * 1.5
change_points = data[data['diff'] > threshold].index.tolist()

# 确保从开始和结束都包含在内
segments = [0] + change_points + [len(data)]

# 对每一段进行平滑处理（使用窗口大小为3的移动平均）
smoothed_values = []
for i in range(len(segments)-1):
    start = segments[i]
    end = segments[i+1]
    segment = data.iloc[start:end]

    # 平滑处理 - 使用移动平均
    smoothed = segment['主路4的车流量 (Traffic flow on the Main road 4)'].rolling(
        window=3, min_periods=1, center=True).mean()

    smoothed_values.extend(smoothed)

# 将平滑后的值添加回原DataFrame
data['smoothed_traffic'] = smoothed_values
plt.figure(figsize=(7,4),dpi=150)
plt.plot(smoothed_values,color='k')
plt.title("平滑后数据",fontsize=13)
plt.xlabel("时刻 (Moment)",fontsize=13)
plt.ylabel("主路4的车流量 (Traffic flow on the Main road 4)",fontsize=11)
plt.grid(True,color='black',alpha=0.25)
plt.tight_layout()  # 紧密布局，防止标题被砍头
plt.show()



print(smoothed_values)
```
