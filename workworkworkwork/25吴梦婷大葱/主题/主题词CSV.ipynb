```python
list = []
with open('tes', 'r') as f:
    for i in range(1,13):
        data = f.readline().split('|')[:-1:]
        each = [i.strip(' ').replace(':',' *') for i in data]
        list.append(each)

sr = pd.DataFrame(data=(i for i in list),index=(f'主题{j}' for j in range(1,13)))
sr.to_csv(path_or_buf='LDA.csv', sep="+")
print(sr)
```
