```python
import pandas as pd
from sympy.polys.polyconfig import query
```

## 1. 直接赋值

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s[:])
print(s)
s[:] = 10
print(s)
```

## 2.通过索引直接对特定位置的值进行替换。

```python
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s['b'] = 20
s[['a', 'e']] = [100, 500]
print(s)
```

## 3.根据条件过滤 Series 并对符合条件的元素赋值。

```python
s = pd.Series([1, 2, 3, 4, 5])
s[s > 2] = 10
print(s)
```

## 4. 使用 .loc 和 .iloc 赋值

```python
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

# 使用 .loc（基于标签）
s.loc['b':'d'] = 20
print(s)

# 使用 .iloc（基于整数位置）
s.iloc[3:5] = 30
print(s)
```

## 5. 按条件表达式赋值

```python
s = pd.Series([1, 2, 3, 4, 5])
s.query((s > 2) & (s < 5))
```

## 6. 使用 .where() 和 .mask() 赋值

```python
s = pd.Series([1, 2, 3, 4, 5])

# where：满足条件的保留，不满足的替换为指定值
s.where(s > 3, 100)  # 条件为 s > 3，否则替换为 100
print(s)

# mask：与 where 相反，满足条件的替换，不满足的保留
s.mask(s > 3, 200)  # 条件为 s > 3，替换为 200
print(s)
```

## 7. 使用 .apply() 进行函数赋值

```python
s = pd.Series([1, 2, 3, 4, 5])
s = s.apply(lambda x: x * 2 if x > 2 else x)
print(s)
```
