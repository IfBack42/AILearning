## 1. 行连接（纵向合并）

- 含义：将两个表格 上下堆叠，增加行数（类似在 Excel 中复制粘贴到下方）。

- 适用场景：合并 结构相同 的数据（如不同月份的数据表）。

- 方import pandas as pd
```
# 示例数据
df1 = pd.DataFrame({"姓名": ["张三", "李四"], "年龄": [25, 30]})
df2 = pd.DataFrame({"姓名": ["王五", "赵六"], "年龄": [28, 35]})

# 行连接（纵向合并）
result = pd.concat([df1, df2], axis=0)法：

```
- 输出结果：
```
  姓名  年龄
0 张三  25
1 李四  30
0 王五  28
1 赵六  35
  ```
- 关键参数：axis=0（默认值，可省略

## 2. 列连接（横向合并）

- 含义：将两个表格 左右拼接，增加列数（类似在 Excel 中复制粘贴到右侧）。

- 适用场景：合并 不同特征 的数据（如用户基本信息表和消费记录表）。

- 方法：
```
# 示例数据
df_info = pd.DataFrame({"姓名": ["张三", "李四"], "性别": ["男", "女"]})
df_score = pd.DataFrame({"语文": [90, 85], "数学": [80, 95]})

# 列连接（横向合并）
result = pd.concat([df_info, df_score], axis=1)
```
- 输出结果：
```
  姓名 性别  语文 数学
0 张三 男  90  80
1 李四 女  85  95
```
- 关键参数：axis=1。

#  对比总结
|操作类型|合并方向|参数	|效果	示例场景|
|----|----|----|---|
|行连接|纵向合并|axis=0|	上下堆叠，增加行数	合并 1 月和 2 月销售数据|
|列连接|横向合并|axis=1|	左右拼接，增加列数	合并用户基本信息和消费记录|


## 注意事项

### 索引重置：合并后行索引可能重复，可用 ignore_index=True 重置：
### 列名一致：
- 行连接时，列名和顺序必须完全一致。
```pd.concat([df1, df2], axis=0, ignore_index=True)```
- 列连接时，行数必须一致（否则填充 NaN）。
