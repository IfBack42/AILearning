```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimSun"]
plt.rcParams["axes.unicode_minus"] = False
```

```python
data = pd.read_csv('../../数模/第一次校赛/data/甲烷.tsv')
data
```

```python
"""
matplot创建基本步骤
1.创建画布
2.准备数据集，x轴y轴
可选：颜色、线型、线宽、标签、网格、图例、标题、坐标轴标签
3.绘制图像
"""
```

```python
# 创建画布
plt.figure(figsize=(10,5),dpi=80) #figsize：宽、高 dpi：分辨率
# 准备数据集 x、y轴
x = [1,2,3,4,5,6,7] #星期
y = [9,19,24,28,38,13,12] #日均温
plt.plot(x,y) # plot() 绘制折线图
# 绘制图像
plt.show()
```

```python
# 准备数据 某城市一小时每分钟温度变化
x = range(60)                                                                                   # tolist & around & round
y = np.around(np.random.uniform(high=25,low=5,size=(1,60)),2).tolist()[0]  # 这一串可以记一下，保留小数以及array转list
# 创建画布   around(...,2):保留 2位小数                             tolist():numpy -> list
plt.figure(figsize=(25,8))
# 绘制图形
plt.plot(x,y)

# 辅助功能
# 1.自定义x、y刻度
x_ticks_label = [f'11:{i}' for i in x]
y_ticks = [i for i in range(5,31)]
plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 2.添加网格儿
plt.grid(True,linestyle='--',color='c',alpha=0.5) #linstyle: '--' '-.' ':' alpha：透明度

# 3.x、y的标签，和标题
plt.xlabel('时间',fontsize=15)
plt.ylabel('温度',fontsize=15)
plt.title("泥马勒戈壁",fontsize=20)

# 保存图形
# plt.savefig('./wocao.png')

# 显示图形
plt.show()
```
