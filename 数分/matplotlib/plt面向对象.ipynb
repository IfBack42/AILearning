```python
import random

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimSun"]
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 一个画板多个坐标系
x = range(0,60)
y_baijing = np.random.uniform(5,18,size=(1,60)).tolist()[0]
y_mohe = np.random.uniform(1,10,size=(1,60)).tolist()[0]
# 创建画布
fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(24,8))
# 绘制
ax[0].plot(x,y_baijing,label='baijing')
ax[1].plot(x,y_mohe,label='mohe')
# 刻度
x_ticks_label = [f'11点{i}分' for i in x]
y_baijing_ticks = [i for i in range(0,25)]
y_mohe_ticks = [i for i in range(0,20)]
y_baijing_ticklabels = [f'{i}C°' for i in range(0,25)]
y_mohe_ticklabels = [f'{i}C°' for i in range(0,20)]

ax[0].set_xticks(x[::5])
ax[0].set_xticklabels(x_ticks_label[::5])
ax[0].set_yticks(y_baijing_ticks[::5],y_baijing_ticklabels[::5])  #不如直接和面向过程版塞到一个函数

ax[1].set_xticks(x[::5],x_ticks_label[::5])
ax[1].set_yticks(y_mohe_ticks[::5],y_mohe_ticklabels[::5])

ax[1].set_title('百京儿',fontsize=20)
ax[1].set_ylabel('wendu',fontsize=15)
ax[1].set_xlabel('shijian',fontsize=15)
ax[1].grid(True,linestyle=':',alpha=1)
ax[1].legend(loc=0)


ax[0].set_title('mohe',fontsize=20)
ax[0].set_ylabel('wendu',fontsize=15)
ax[0].set_xlabel('shijian',fontsize=15)
ax[0].grid(True,linestyle=':',alpha=1)
ax[0].legend(loc=0)

plt.show()
```
